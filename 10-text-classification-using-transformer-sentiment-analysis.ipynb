```python
# Step 1: Install required libraries
!pip install transformers torch

# Step 2: Import libraries
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Step 3: Load a pre-trained transformer model and tokenizer (e.g., DistilBERT)
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Step 4: Initialize a text classification pipeline
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Step 5: Test the classifier with example sentences
sentences = [
    "I love learning about transformers!",
    "This is a difficult topic.",
    "Transformers are revolutionizing NLP."
]

# Run the classifier and print the results
results = classifier(sentences)
for sentence, result in zip(sentences, results):
    print(f"Sentence: '{sentence}'\nSentiment: {result['label']}, Score: {result['score']:.2f}\n")
```
