## Linear Regression Coefficients Calculation
This notebook calculates the slope (m) and intercept (c) for a linear model using the direct linear regression formulas.
The formula for the model is: $y = mx + c$.

### Instructions:
1. The data for this notebook is already hosted online.
2. Run the code cells below to calculate the linear regression coefficients and plot the regression line.


```python

# Import necessary libraries
import numpy as np
import pandas as pd

# URL for the dataset
url = 'https://raw.githubusercontent.com/apownukepcc/MLExamples/refs/heads/main/Lab01.csv'

# Load the CSV file from the URL into a DataFrame
data = pd.read_csv(url, header=None, names=['x', 'y'])

# Show the first few rows of the data
data.head()
```

```python

# Extract x and y values from the data
x = data['x']
y = data['y']

# Compute the means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate the terms needed for the numerator and denominator of the slope (m)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

# Calculate slope (m)
m = numerator / denominator

# Calculate intercept (c)
c = y_mean - m * x_mean

# Output the slope and intercept
print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Optionally, predict values using the obtained model
y_pred = m * x + c

# Output predicted values
y_pred
```

```python

# Import necessary libraries for plotting
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(x, y, color='blue', label='Data points')

# Plot the regression line
plt.plot(x, y_pred, color='red', label='Regression line')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
